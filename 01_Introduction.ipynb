## 1. Giới thiệu (Introduction)
Netflix hiện là nền tảng phát trực tuyến lớn nhất thế giới với thư viện nội dung khổng lồ. Tuy nhiên, việc hiểu rõ chiến lược phân phối nội dung và giúp người dùng tìm kiếm phim phù hợp giữa "biển" dữ liệu này là một thách thức lớn.

Dự án này sử dụng bộ dữ liệu **Netflix Movies and TV Shows** để giải quyết hai mục tiêu chính:
1.  **Phân tích chiến lược (Business Intelligence):** Khám phá xu hướng sản xuất theo quốc gia, đối tượng khán giả (Rating) và tính mùa vụ trong thời điểm phát hành.
2.  **Ứng dụng kỹ thuật (Machine Learning):** Xây dựng hệ thống gợi ý phim (Recommendation System) dựa trên nội dung để cá nhân hóa trải nghiệm người dùng.

**Quy trình thực hiện:**
1.  **Data Collection:** Thu thập và tìm hiểu dữ liệu.
2.  **Data exploration** Khám phá dữ liệu
3.  **Preprocessing:** Làm sạch dữ liệu (xử lý missing values, chuẩn hóa ngày tháng, tách chuỗi).
4.  **Data Analysis (EDA):** Trực quan hóa để trả lời 3 câu hỏi chiến lược.
5.  **Machine Learning:** Xây dựng Content-Based Recommendation System sử dụng TF-IDF.
6.  **Project Summary:** Tóm tắt dự án.



## 2. Thu thập dữ liệu
### Chủ đề dữ liệu là gì?
Netflix là một trong những nền tảng truyền thông và phát trực tuyến video phổ biến nhất. Tính đến giữa năm 2021, họ có hơn 8000 phim và chương trình truyền hình trên nền tảng của mình, và hơn 200 triệu người đăng ký trên toàn cầu. Bộ dữ liệu dạng bảng này bao gồm danh sách tất cả các phim và chương trình truyền hình có sẵn trên Netflix, cùng với các chi tiết như: diễn viên, đạo diễn, xếp hạng, năm phát hành, thời lượng, v.v.
### Nguồn gốc?
- **Nguồn:** Kaggle (Netflix Movies and TV Shows).
- **Link:** https://www.kaggle.com/datasets/shivamb/netflix-shows/data
- **Kích thước:** Hơn 8,800 dòng và 12 cột.
- **Loại dữ liệu:** Đa dạng (Văn bản, Thời gian, Phân loại).
- **Tác giả** Shivamb Bansal (also listed as Shivam Bansal)
- **Ngày xuất bản** Phiên bản mới nhất được phát hành ngày 26 tháng 9 năm 2021 (phiên bản gốc khoảng năm 2020).
### Giấy phép sử dụng?
Dataset này có giấy phép CC0: Public Domain, nghĩa là được phép dùng tự do không cần ghi tên tác giả hay lo lắng về vấn đề bản quyền.

Vì vậy dataset này hoàn toàn có thể được sử dụng cho mục đích giáo dục
### Dữ liệu này được thu thập thế nào?
- Dữ liệu được thu thập qua web scraping từ Flixable, một công cụ tìm kiếm bên thứ ba cho Netflix.​
- Đối tượng là toàn bộ phim và chương trình TV có sẵn trên Netflix (chủ yếu thị trường Mỹ), không sử dụng lấy mẫu ngẫu nhiên mà thu thập toàn bộ danh sách lúc đó.​
- Thời gian thu thập khoảng năm 2019, với dữ liệu phản ánh nội dung Netflix đến thời điểm đó.​
- Hạn chế: Dữ liệu có thể lỗi thời (không cập nhật thêm phim mới), thiên bias theo khu vực (chủ yếu Mỹ), và thiếu thông tin chi tiết như nguồn gốc chính xác hoặc dữ liệu người dùng thực tế.​
### Tại sao chọn bộ dữ liệu này?
- Bộ dữ liệu Netflix Movies and TV Shows được chọn vì tính sẵn có cao trên Kaggle và kích thước vừa phải (khoảng 8.800 phim/show), phù hợp cho phân tích khám phá dữ liệu giáo dục.
- Phim ảnh cũng là chủ đề yêu thích của nhóm.
